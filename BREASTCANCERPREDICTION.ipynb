```python
dfn['breast'] = dfn['breast'].map({1: 0, 2: 1})
```

```python
columns_to_encode = ['age', 'inv-nodes', 'deg-malig', 'tumor-size', 'irradiat', 'breast-quad', 'menopause']
df_encoded = pd.get_dummies(dfn, columns=columns_to_encode)

# Convert only the columns created from `columns_to_encode` to int
encoded_columns = df_encoded.columns[df_encoded.columns.str.contains('|'.join(columns_to_encode))]
df_encoded[encoded_columns] = df_encoded[encoded_columns].astype(int)
```

```python
import pandas as pd
from scipy.stats import pearsonr

# Define ranges for power ratios of each variable
tumor_size_powers = range(1, 10)
deg_malig_powers = range(1, 10)
age_powers = range(1, 10)
menopause_powers = range(1, 10)

# Initialize variables to store the best power combinations and correlation
best_tumor_size_power = 0
best_deg_malig_power = 0
best_age_power = 0
best_menopause_power = 0
best_correlation = -1

# Iterate through different power combinations for each variable
for tumor_size_power in tumor_size_powers:
    for deg_malig_power in deg_malig_powers:
        for age_power in age_powers:
            for menopause_power in menopause_powers:
                # Apply individual powers to each variable
                dfn['new_column'] = ((dfn['tumor-size'] ** tumor_size_power) *
                                     (dfn['deg-malig'] ** deg_malig_power)) / (
                                     (dfn['age'] ** age_power) *
                                     (dfn['menopause'] ** menopause_power))

                # Calculate the correlation between 'new_column' and the target 'class'
                correlation = pearsonr(dfn['new_column'], dfn['class'])[0]

                # Update best powers if a higher correlation is found
                if correlation > best_correlation:
                    best_tumor_size_power = tumor_size_power
                    best_deg_malig_power = deg_malig_power
                    best_age_power = age_power
                    best_menopause_power = menopause_power
                    best_correlation = correlation

# Apply the best power combinations to calculate the final 'new_column'
dfn['new_column'] = ((dfn['tumor-size'] ** best_tumor_size_power) *
                     (dfn['deg-malig'] ** best_deg_malig_power)) / (
                     (dfn['age'] ** best_age_power) *
                     (dfn['menopause'] ** best_menopause_power))

# Print the results
print("Best tumor-size power:", best_tumor_size_power)
print("Best deg-malig power:", best_deg_malig_power)
print("Best age power:", best_age_power)
print("Best menopause power:", best_menopause_power)
print("Best correlation:", best_correlation)
```

```python
import numpy as np
from scipy.stats import pearsonr

# Given medians for each class
median_0 = 22.204082
median_1 = 119.73913

# Initialize variables to store the best threshold and correlation for each class
best_threshold_0 = 0
best_threshold_1 = 0
best_correlation_0 = -1
best_correlation_1 = -1

# Define a range of thresholds around the medians for each class for testing
threshold_range_0 = np.linspace(median_0 * 0.5, median_0 * 1.5, 100)  # Adjusting range around median_0
threshold_range_1 = np.linspace(median_1 * 0.5, median_1 * 1.5, 100)  # Adjusting range around median_1

# Brute-force search for the best threshold for class 0
for threshold in threshold_range_0:
    # Apply the threshold for class 0
    dfn['binary_column'] = dfn['new_column'].apply(lambda x: 1 if x >= threshold else 0)

    # Calculate correlation with the 'class' target column
    correlation = pearsonr(dfn['binary_column'], dfn['class'])[0]

    # Update best threshold for class 0 if a higher correlation is found
    if correlation > best_correlation_0:
        best_threshold_0 = threshold
        best_correlation_0 = correlation

# Brute-force search for the best threshold for class 1
for threshold in threshold_range_1:
    # Apply the threshold for class 1
    dfn['binary_column'] = dfn['new_column'].apply(lambda x: 1 if x >= threshold else 0)

    # Calculate correlation with the 'class' target column
    correlation = pearsonr(dfn['binary_column'], dfn['class'])[0]

    # Update best threshold for class 1 if a higher correlation is found
    if correlation > best_correlation_1:
        best_threshold_1 = threshold
        best_correlation_1 = correlation

# Apply the best threshold to create the final 'binary_column'
# You may combine thresholds as needed for practical usage, e.g., averaging, to get a single threshold.
best_threshold = (best_threshold_0 + best_threshold_1) / 2
dfn['binary_column'] = dfn['new_column'].apply(lambda x: 1 if x >= best_threshold else 0)

# Print the results
print("Best threshold for class 0:", best_threshold_0)
print("Best correlation for class 0:", best_correlation_0)
print("Best threshold for class 1:", best_threshold_1)
print("Best correlation for class 1:", best_correlation_1)
print("Final combined threshold:", best_threshold)
```
